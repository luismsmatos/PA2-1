```python
# Importa as "libraries" necessárias para executar o programa
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import seaborn as sns
sns.set_style("white")
import pandas as pd
my_dpi=96
```

```python
# Vai buscar o ficheiro .csv que está alojado na web e mostra as 5 primeiras linhas do ficheiro
url = 'https://python-graph-gallery.com/wp-content/uploads/gapminderData.csv'
data = pd.read_csv(url)
#data = pd.read_csv('gapminderData.csv')
data.head()
```

```python
# Transforma a coluna "Continent" em valores numéricos
data['continent']=pd.Categorical(data['continent'])
```

```python
%matplotlib notebook
%matplotlib notebook
 
# É un loop que vai buscar os anos para os gráficos
for i in data.year.unique():
 
# Inicia a figura
    fig=plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)
 
# Muda as cores com c e alpha. Depois é mapeado o valor para o eixo do X
    tmp = data[ data.year == i ]
    plt.scatter(tmp['lifeExp'], tmp['gdpPercap'] , s=tmp['pop']/200000 , c=tmp['continent'].cat.codes, cmap="Accent", alpha=0.6, edgecolors="white", linewidth=2)
 
# Adiciona os titulos ao eixo do X e do Y. Limita também os valores destes mesmos eixos
    plt.yscale('log')
    plt.xlabel("Life Expectancy")
    plt.ylabel("GDP per Capita")
    plt.title("Year: "+str(i) )
    plt.ylim(0,100000)
    plt.xlim(30, 90)
 
# Guarda as imagens como "Gapminder_step + ano + .png"
    filename='Gapminder_step'+str(i)+'.png'
    plt.savefig(filename, dpi=96)
    plt.gca()

#Transformation in a GIF using Image Magick

# Then use image magick (this is bash, not python)
#magick convert -delay 80 Gapminder*.png animated_gapminder.gif
```

```python
# Liste a tarifa média paga pelos passageiros em cada uma das cidades de embarque.
f = open('D://transferecias chrome/Titanic.csv')
titanicDF = pd.read_csv(f)
titanicDF
#media da cidade C
df= titanicDF[(titanicDF['Embarked'] == 'C')]
MediaC=df['Fare'].mean()
#media da cidade Q
df2= titanicDF[(titanicDF['Embarked'] == 'Q')]
MediaQ=df2['Fare'].mean()
#media da cidade S
df3= titanicDF[(titanicDF['Embarked'] == 'S')]
MediaS=df3['Fare'].mean()
print("C:",MediaC," Q:",MediaQ," S:",MediaS)
```

```python
# Os valores inexistentes (missing)  no Pandas são nulos (null), impressos como 'NaN' e
# a function pd.isnull(v) verifica se um valor é nulo.
# Que fração dos passageiros não tem valor para a idade?
f = open('D://transferecias chrome/Titanic.csv')
titanicDF = pd.read_csv(f)
titanicDF
#crio um dataframe a partir do titanic com os valores da idade que estao nulos
df= titanicDF[(titanicDF['Age'].isnull())]
#faço a divisao entre o dataframe criado e o original para descobrir a percentagem entre os que têm a idade a nulo e o total
fracao= len(df)/len(titanicDF)
print("{:.2%}".format(fracao))
```
